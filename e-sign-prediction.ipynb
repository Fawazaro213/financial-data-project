## Feature Engineering

```python
e_sign_1["years_employed"] = e_sign_1["years_employed"] * 12 
e_sign_1["month_employed"] = e_sign_1["years_employed"] +  e_sign_1["months_employed"]

e_sign_1.head()
```

```python
e_sign_1.drop("months_employed", axis=1, inplace=True)
e_sign_1.drop("years_employed", axis=1, inplace=True)
```

```python
e_sign_1.head()
```

```python
e_sign_1["personal_account_y"] = e_sign_1["personal_account_y"] * 12 
e_sign_1["personal_account"] = e_sign_1["personal_account_y"] +  e_sign_1["personal_account_m"]

e_sign_1.head()
```

```python
e_sign_1.drop("personal_account_y", axis=1, inplace=True)
e_sign_1.drop("personal_account_m", axis=1, inplace=True)
```
